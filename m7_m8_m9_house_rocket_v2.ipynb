## 1.1 Helper Fuctions

```python
def grafdisplot(dt):
    sns.set(rc={'figure.figsize':(20, 10)})
    sns.distplot(dt, kde=True);
    
    return None


def boxplot1(x1, x2, x3, x4, y, df):
    
    plt.subplot(2, 2, 1)
    sns.boxplot(x=x1, y=y, data=df)
    
    plt.subplot(2, 2, 2)
    sns.boxplot(x=x2, y=y, data=df)

    plt.subplot(2, 2, 3)
    sns.boxplot(x=x3, y=y, data=df)

    plt.subplot(2, 2, 4)
    sns.boxplot(x=x4, y=y, data=df);

    return None


def categoricalvar(co1, c1, l1, c2, l2, c3, l3, c4, l4,
                   co2, c5, l5, c6, l6, c7, l7,
                   co3, c8, l8, c9, l9, c10, l10,
                   co4, c11, l11, c12, l12): 
    # level
    plt.figure(figsize=(25, 20))
    plt.subplot(4, 2, 1)
    sns.countplot(x=co1)

    plt.subplot(4, 2, 2)
    sns.kdeplot(x=c1, label=l1, shade=True)
    sns.kdeplot(x=c2, label=l2, shade=True)
    sns.kdeplot(x=c3, label=l3, shade=True)
    sns.kdeplot(x=c4, label=l4, shade=True)
    plt.legend()

    # dormitory_type
    plt.subplot(4, 2, 3)
    sns.countplot(x=co2)

    plt.subplot(4, 2, 4)
    sns.kdeplot(x=c5, label=l5, shade=True)
    sns.kdeplot(x=c6, label=l6, shade=True)
    sns.kdeplot(x=c7, label=l7, shade=True)
    plt.legend()
    
    # condition_type
    plt.subplot(4, 2, 5)
    sns.countplot(x=co3)

    plt.subplot(4, 2, 6)
    sns.kdeplot(x=c8, label=l8, shade=True)
    sns.kdeplot(x=c9, label=l9, shade=True)
    sns.kdeplot(x=c10, label=l10, shade=True)
    plt.legend()

    # data_age
    plt.subplot(4, 2, 7)
    sns.countplot(x=co4)

    plt.subplot(4, 2, 8)
    sns.kdeplot(x=c11, label=l11, shade=True)
    sns.kdeplot(x=c12, label=l12, shade=True)
    plt.legend();
    
    return None

def cramer_v(x, y):
    cm = pd.crosstab(x, y).values
    n = cm.sum()
    r, k = cm.shape

    chi2 = ss.chi2_contingency(cm)[0]
    chi2corr = max(0, chi2 - (k - 1) * (r - 1) / (n - 1))
    
    kcorr = k - (k - 1) ** 2 / (n - 1)
    rcorr = r - (r - 1) ** 2 / (n - 1)
    
    return np.sqrt((chi2corr / n) / (min(kcorr - 1, rcorr - 1)))
```

## 8.1 Quais casas a House Rocket deveria comprar e por qual preço de compra?

- As casas que a House Rocket deveria comprar devem está abaixo da mediana. As casa com vista para a água deve ser avaliadas individualmente.

```python
# casas abaixo da mediana
houses_buy = df8[df8['price'] <= df8['price'].median()]

print('São {} casas que tem o preço abaixo da mediana.'.format(len(houses_buy)))
```

```python
# casas que estajam nas condições good e regular
h_buy_cond_type = houses_buy[(houses_buy['condition_type'] == 'good')]

print('São {} casas que tem o preço abaixo da mediana que estejam na condição good.'
      .format(len(h_buy_cond_type)))
```

```python
# casas com vista para a água que o preço estaja menor que o dobro da mediana
hwf_bolew1 = df8[(df8['%_above_median'] > 0) & (df8['%_above_median'] < 100.0)]
print('Quantidade de casas com vista para a água que não são 100% mais caras que a mediana do preço: {} unidades'
      .format(len(hwf_bolew1)))
```

```python
# salvar dataset das casa com vista para a água com o preço menor que o dobro da mediana
hwf_bolew1.to_csv('data/house_waterfront_upmedian.csv', encoding='utf-8', index=False)
```

## 8.2 A House Rocket deveria fazer uma reforma para aumentar o preço da venda?
- As casa com condições bad deveriam passar por uma reforma, já as regulares deveriam ser verificadas caso a cado. As casas com vista para a água deveria passar por uma reforma para ser mais atrativas.

```python
# casas para reforma
h_buy_cond_type2 = houses_buy[(houses_buy['condition_type'] == 'regular') | 
                              (houses_buy['condition_type'] == 'bad')]

print('São {} casas que tem o preço abaixo da mediana que estejam na condição bad ou regular.'
      .format(len(h_buy_cond_type2)))

h_buy_cond_type3 = houses_buy[(houses_buy['condition_type'] == 'bad')]                          
h_buy_cond_type4 = houses_buy[(houses_buy['condition_type'] == 'regular')]

print('São {} casas que tem o preço abaixo da mediana que estejam na condição bad que precisam de uma reforma.'
      .format(len(h_buy_cond_type3)))
print('São {} casas que tem o preço abaixo da mediana que estejam na condição regular que precisam ser avaliadas '
      'caso a caso.'
      .format(len(h_buy_cond_type4)))
```

```python
# casas regulares com vista para a água abaixo da mediana
h_buy_cond_type5 = houses_buy[(houses_buy['condition_type'] == 'regular') & (houses_buy['waterfront'] == 1)]
print('São {} casas que estejam na condição regular abaixo da mediana com vista para a água que devem ter reformas.'
      .format(len(h_buy_cond_type5)))

print('\n')

# casas com vista para o mar acima da medianan e menor que o dobro da mediana no preço
hwf_count = hwf_bolew1[['id', 'condition_type']].groupby('condition_type').count().reset_index()

print('São casas que esteja acima da mediana e que não custe o do dobro do preço com vista para a água que devem ter reformas')

print(hwf_count)

print('As casas que tem a condição bad deve passar por reformas, as regulares devem ser avaliadas caso a caso.')
```

## 8.3 Qual o incremento no preço dado por cada opção de reforma?
- Casas com condições bad devem ter um custo de até 20% no valor na reforma.
- Casas com condições regular devem ter um custo de até 12% no valor na reforma.
    - As casas podem ter um acrécimo ou diminuição nessa porcentagem de acordo com o grau de qualidade do imóvel

```python
# concat house to buy
houses_buy_01 =  pd.concat([houses_buy, hwf_bolew1], axis=0)
```

- Tabela de opções de preço para cada condição de reforma:
    - Casas com condições bad: 20% do valor da casa
    - Casa com condições regular: 12% do valor da casa


### 8.3.1 Adicionar tabela de preço com acrescimo da reforma

```python
# reset o index
houses_buy_01 = houses_buy_01.reset_index().drop(columns='index')
```

```python
# adcionar o custo da reform ano preço 
houses_buy_01['price+renovation'] = houses_buy_01[['condition_type', 'price']].apply(lambda x: x['price'] * 1.20 if x['condition_type'] == 'bad' 
                                                                                     else x['price'] * 1.12 if x['condition_type'] == 'regular' else 
                                                                                     x['price'], axis=1)

# criar coluna so com os preços estimado da reforam
houses_buy_01['renovation_cost'] = houses_buy_01[['condition_type', 'price', 'price+renovation']].apply(lambda x: 
                                                                                                        x['price+renovation'] - x['price'] 
                                                                                                        if x['condition_type'] == 'bad' 
                                                                                                        else x['price+renovation'] - x['price'] 
                                                                                                        if x['condition_type'] == 'regular' else 0.0, axis=1)
                                                                                     
```

```python
houses_buy_01.to_csv('data/houses_buy.csv', encoding='utf-8', index=False)
```
